```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y=x**2
```

```python
y
```

```python
plt.plot(x,y,'lightgreen')
plt.title("Este es mi grafico")
plt.xlabel("Eje x")
plt.ylabel("Ejer y")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'g')

plt.subplot(2,1,2)
plt.plot(x,y,'blue')
```

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.8,0.8])
graficos.plot(x,y,'g')
```

```python
figura  = plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')

grafico2 =  figura.add_axes([0.15,0.5,0.3,0.3])
grafico2.plot(x,y,'g')
```

# Multigraficos

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)

graficos[0][0].plot(x,y,'blue')
graficos[0][1].plot(x,y,'red')
graficos[1][0].plot(x,y,'yellow')
graficos[1][1].plot(x,y,'g')

graficos[0][0].set_title('Azul')
graficos[0][1].set_title('Rojo')
graficos[1][0].set_title('Amarillo')
graficos[1][1].set_title('Verde')
```

# Tamaño grafico

```python
figura = plt.figure(figsize=(8,4))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
```

# Dos graficos en la misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2,label='Al cuadrado')
graficos.plot(x,x**3,label='Al cubo')
graficos.legend(loc=0)
```

# Color del grafico

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,color='red')
```

# Linea del grafico

```python
figura = plt.figure()
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'red',linewidth=5, alpha=0.5, linestyle=':')
```
